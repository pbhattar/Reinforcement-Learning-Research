```python
#takes a column vector, of 0s and exactly one 1, representing a state and returns the index of 1 in this state
def getStateNum(matrix):
    list = matrix.transpose().tolist()[0]
    return list.index(1)

#converts state num to list of 0s and exactly one 1
def getStateList(num, total):
    return ([0] * num) + [1] + ([0] * (total - num -1))   

#can replace this with the built-in function in numpy
#def vecMult(v1, v2):
#    v1 = v1.transpose().tolist()[0]
#    v2 = v2.transpose().tolist()[0]
#    new = [float(a)*b for a,b in zip(v1,v2)]
#    new = np.matrix(new).transpose()
#    return new

#multiplies every row of "matrix" with the row "vector"
def matrixVecMult(matrix, vector): #takes rowVector as parameter
    vectorList = vector.tolist()[0]
    matrixList = matrix.tolist()
    newMatrixList = []
    for row in matrixList:
        new = [float(a)*b for a,b in zip(vectorList,row)]
        newMatrixList.append(new)
    newMatrix = np.matrix(newMatrixList)
    return newMatrix
        
#for debugging purposes; returns the correct (s|s') distribution    
def getCorrectStatesCPT(states_P, numRealStates):
    state_absState = []
    pairs = states_P.items()
    numAbsStates = len(pairs)
    m = np.zeros((numRealStates, numAbsStates))
    for oneItem in pairs:
        absState =  oneItem[0]
        absStateList = getStateList(absState, numAbsStates)
        for realState in oneItem[1]:
            realStateIndex = getStateNum(realState)
            m[realStateIndex] = absStateList
    return np.matrix(m)

#returns the (a'|s') CPT; CPT[i,j] = 0 means that action j is not available in state i
'''all entries add upto 1, is that correct?? I also treat this as priors for my beliefs'''
def getUniformAbsActAbsStateMatrix(numAbsStates, numAbsActions,absMDP):
    cptLists = []
    colNormalizeConstant = 1/float(numAbsStates)
    for s in range(numAbsStates):
        row = []
        actionsAvailable = absMDP.getActions(np.matrix(getStateList(s, numAbsStates)).transpose())
        rowNormalizeConstant = 1/float(len(actionsAvailable))
        for a in range(numAbsActions):
            if a in actionsAvailable:
                row.append(rowNormalizeConstant * colNormalizeConstant)
            else:
                row.append(0.0)
        cptLists.append(row)
    cpt = np.matrix(cptLists) # rows are absStates, items are P(s|s'), or P(a'|s')
    return cpt

#returns uniform CPT where s|s'_i = s|s'_j for all s, i and j
def getUniformCPTMatrix(numRStates, numAbsStates):
    cptLists = []
    for i in range(numRStates): 
        row = [1/float(numAbsStates)] * numAbsStates
        cptLists.append(row)
    cpt = np.matrix(cptLists) # rows are absStates, items are P(s|s'), or P(a'|s')
    return cpt


#def getNextStateDistribution(actionMatrix, absState):
#    return (actionMatrix[absState]).transpose()

def normalizeVector(v): #takes a vector, row or column
    s = np.sum(v)
    if s!= 0:
        v = (float(1)/s) * v
    return v  #returns a vector

def normalizeMatrix(a):
    s = np.sum(a)
    new_matrix = a / float(s)
    return new_matrix

def normalizeList(l):
    total = sum(l)
    newL = []
    for num in l:
        newL.append(num/float(total))
    return newL

#given a transition matrix and a startState, sample a nextState
def sample(matrix, start):
    #print "action ", matrix
    row = matrix[start].tolist()[0]
    #print "row ", row
    t = np.random.multinomial(1, row).tolist()
    nextState = t.index(1)
    return nextState

def terminalState(stateNum):
    #if stateNum == 2 or stateNum == 5 or stateNum == 8:
    if stateNum == 9:
        return True
    else:
        return False

#given a row vector with entries corresponding to reward in the states in an MDP and an observedReward, 
#returns a row vector with entries corresponding to (observedReward|state)    
def getRewardDistribution(absRVec, obsReward):
    #absRVec is a row vector
    tempRVec = np.matrix(absRVec)
    for i in range(tempRVec.shape[1]):
        if tempRVec[0,i] == obsReward:
            tempRVec[0,i] = 1
        else:
            tempRVec[0,i] = 0
    tempRVec = normalizeVector(tempRVec)
    return tempRVec
```

```python
#returns and episode of the form ((state, reward, action), state, reward, action, .....state, reward)
def getEpisodeInTuples(mdp, steps):
    episode = [] 
    states = mdp.getStates()
    mdpReward = mdp.getReward()
    startStateNum = random.choice(range(len(states)))
    stepNum = 0
    while stepNum != steps and not terminalState(startStateNum):
        onePiece = [startStateNum]
        actions = mdp.getActions(np.matrix(h.makeStateList(startStateNum, len(states))).transpose())
        takeAction = random.choice(actions)
        rewardInTuple = mdpReward[0, startStateNum]
        onePiece.append(rewardInTuple)
        onePiece.append(takeAction)
        #print "action ", takeAction
        nextStateNum = sample(mdp.getTransitionProbabilities(takeAction), startStateNum)
        startStateNum = nextStateNum
        stepNum += 1
        episode.append(tuple(onePiece))
    return tuple(episode)
```

```python
#calculates backward beliefs; returns a matrix that has been epsilon adjusted and normalized
def getBackwardBeliefs(absMDP, tupleEpisode, epsilon, cpt_absAct_givenAbsState, dict_stateAction_givenAbsAction, cpt_state_givenAbsState,absRewardVec):
    nodes = len(tupleEpisode)
    numAbsStates = len(absMDP.getStates()) #calculate again or just pass? Pass too many variables?
    numAbsActions = len(absMDP.getAllActions())
    backwardBeliefs = {} 
    backwardBeliefs[nodes] = cpt_absAct_givenAbsState
    for t in range(nodes-1,-1,-1): #goes from nodes-1 to 0 [nodes-1 is the last episode]
        oneTuple = tupleEpisode[t]
        observedState, observedReward, observedAction = oneTuple[0], oneTuple[1], oneTuple[2]
        stateAction_givenAbsAction = dict_stateAction_givenAbsAction[(observedState, observedAction)] 
        #row vector of size numAbsActions with entries corresponding to s,a|a'
        rewardGivenAbsState = getRewardDistribution(absRewardVec, observedReward) #row vector
        #row vector of size numAbsStates with entries corresponding to observedReward|s'
        state_givenAbsState = cpt_state_givenAbsState[observedState] #row vector
        #row vector corresponding of size numAbsStates with entries corresponding to observedState|s'
        backwardBelief_t = np.matrix(np.zeros((numAbsStates,numAbsActions)))
        for i in range(numAbsStates):  
            for j in range(numAbsActions): 
                transitionMatrix = absMDP.getTransitionProbabilities(j)
                transitionFrom_s_i = transitionMatrix[i]
                belief_ij = 0
                for a_l in range(numAbsActions): 
                    stateAction_givenAbsAction_a_l = stateAction_givenAbsAction[0,a_l]
                    for s_k in range(numAbsStates): 
                        observationConditionals = rewardGivenAbsState[0,s_k] * stateAction_givenAbsAction_a_l * state_givenAbsState[0,s_k] * backwardBeliefs[t+1][s_k,a_l]
                        transitionConditionals = transitionFrom_s_i[0,s_k] 
                        belief_ij += (observationConditionals * transitionConditionals)
                backwardBelief_t[i,j] = belief_ij
        backwardBelief_t = adjustEpsilon(backwardBelief_t, epsilon) #should this be normalized?
        backwardBelief_t = normalizeMatrix(backwardBelief_t)
        backwardBeliefs[t] = backwardBelief_t
    return backwardBeliefs
    
```

```python
#given a belief(2 dimensional matrix) for abstract state and abstract actions, return (1 dimensional vectors)
#consisting of beliefs s|s' and s,a|a' 
def getBeliefAlongAxis(beliefMatrix):
    beliefStates = {}
    beliefActions = {}
    for key in beliefMatrix:
        beliefForTuple = beliefMatrix[key]
        beliefActions[key] = np.sum(beliefForTuple, axis=0)
        beliefStates[key] = (np.sum(beliefForTuple, axis=1)).transpose()
    return beliefStates, beliefActions
```

```python
#maintain count of (state, abstractState) and abstractState
def makeCountsForStates(belief, episode,cComb,cAbs):
    for i in range(len(episode)):
        beliefV = belief[i]
        state = episode[i][0]
        for absNum in range(beliefV.shape[1]):
            cComb[(state, absNum)] += beliefV[0,absNum]
            cAbs[absNum] += beliefV[0, absNum]
    return cComb, cAbs
```

```python
#maintain count of (abstractAction, state, actiontState) and abstractAction
def makeCountsForActions(belief, episode,cTriple,cAbsAction):
    for i in range(len(episode)):
        beliefV = belief[i]
        state = episode[i][0]
        action = episode[i][2]
        for absActionNum in range(beliefV.shape[1]):
            cTriple[(absActionNum,state,action)] += beliefV[0,absActionNum]
            cAbsAction[absActionNum] += beliefV[0,absActionNum]
    return cTriple, cAbsAction
```

```python
#use the cummulative counts to update s|s' CPT
def updateStatesCPT(countComb, countAbs, numRStates, numAbsStates): #dont need visitedStates
    newCPT = []
    for stateNum in range(numRStates):
        stateProbs = []
        for absStateNum in range(numAbsStates):
            numerator =  countComb[(stateNum, absStateNum)] + 0.001
            denominator = float(countAbs[absStateNum]) + 0.1
            prob_stateGivenAbsState = numerator / float(denominator)
            stateProbs.append(prob_stateGivenAbsState)
        stateProbs = normalizeList(stateProbs)
        newCPT.append(stateProbs)
    newCPT = np.matrix(newCPT)
    return newCPT 
```

```python
#use the cummulative counts to update s,a|a' CPT (dictionary)
def updateActionsCPT(countTriples, countAbsActions):
    new_dict = util.Counter()
    numAbsActions = len(countAbsActions.keys()) 
    for tup in dict_stateAction_givenAbsAction.keys():
        row = []
        for absAction in range(numAbsActions):
            stateNum,actionNum = tup[0],tup[1]
            numerator =  countTriples[(absAction,stateNum, actionNum)] + 0.001
            denominator = float(countAbsActions[absAction]) + 0.1
            row.append(numerator / float(denominator))
        new_dict[tup] = np.matrix(row)
    return new_dict
```
